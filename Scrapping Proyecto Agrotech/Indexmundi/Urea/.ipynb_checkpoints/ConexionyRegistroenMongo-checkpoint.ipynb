### Propósito de este scraper y página objetivo:

- El objetivo es scrapear la siguiente página Indexmundi que contiene los datos de los precios de productos agrícolas y materias primas.

Link: https://www.indexmundi.com/es/precios-de-mercado/?mercancia=cloruro-de-potasio&meses=240&moneda=clp

- En este caso, se extrajo mes, año, precio y variación porcentual del valor de la tonelada de fosfato diamónico (fertilizante) en peso chileno

```python
from pymongo import MongoClient

# 1. Conectar al contenedor de Mongo
client = MongoClient('mongodb://mongodb:27017/')

# 2. Crear (o usar) la base de datos y la colección
db = client['previa_proyecto']
coleccion = db['fosfato_diamonico']

print("Conexión exitosa a MongoDB")
```

```python
import requests
from bs4 import BeautifulSoup
import time
import re
from pymongo import MongoClient
from collections import OrderedDict

# --- CONFIGURACIÓN ---
NOMBRE_GRUPO = "G9-AgroTech"
NOMBRE_INTEGRANTE = "Sebastián Castillo"
NOMBRE_ETIQUETA = "Fosfato diamónico"
# URL ajustada para asegurar que el rango cubra desde 2020 hasta el presente
URL_OBJETIVO = "https://www.indexmundi.com/es/precios-de-mercado/?mercancia=fosfato-diamonico&meses=360&moneda=clp"

meses_completos = {
    "ene.": "enero", "feb.": "febrero", "mar.": "marzo", "abr.": "abril",
    "may.": "mayo", "jun.": "junio", "jul.": "julio", "ago.": "agosto",
    "sep.": "septiembre", "oct.": "octubre", "nov.": "noviembre", "dic.": "diciembre"
}

# --- PASO 1: SCRAPING Y FILTRADO (Desde 2020) ---
datos_finales = []
try:
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    response = requests.get(URL_OBJETIVO, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Selector flexible para evitar perder meses recientes
    filas = soup.find_all('tr')
    
    print(f"--- 1. Iniciando extracción de {NOMBRE_ETIQUETA} ---")
    
    for fila in filas:
        celdas = fila.find_all(['td', 'th'])
        if len(celdas) >= 2:
            texto_fila = [c.get_text(strip=True) for c in celdas]
            
            # Regex para capturar patrón: [mes.] [año]
            match_fecha = re.search(r'([a-z]{3}\.?)\s+(\d{4})', texto_fila[0].lower())
            
            if match_fecha:
                abr_mes = match_fecha.group(1)
                if not abr_mes.endswith('.'): abr_mes += '.'
                anio_str = match_fecha.group(2)
                anio_int = int(anio_str)
                
                # FILTRO: Enero 2020 en adelante
                if anio_int >= 2020:
                    precio_raw = texto_fila[1]
                    # Si no hay columna de variación (guion), se captura como "-"
                    variacion_raw = texto_fila[2] if len(texto_fila) > 2 else "-"
                    
                    if precio_raw and any(char.isdigit() for char in precio_raw):
                        nombre_mes = meses_completos.get(abr_mes, abr_mes)
                        
                        item = OrderedDict()
                        item["grupo"] = NOMBRE_GRUPO
                        item["Integrante"] = NOMBRE_INTEGRANTE
                        item["Etiqueta"] = NOMBRE_ETIQUETA
                        item["Año"] = anio_str
                        item["Mes"] = nombre_mes
                        item["valor_sucio"] = precio_raw
                        item["variacion_sucia"] = variacion_raw
                        datos_finales.append(item)

    print(f"✅ Extracción exitosa: {len(datos_finales)} registros encontrados.")
except Exception as e:
    print(f"❌ Error en Scraping: {e}")

# --- PASO 2: CARGA A MONGODB ---
if datos_finales:
    # Usamos el nombre del host verificado en tu Docker: bigdata_mongodb
    host_docker = 'mongodb://bigdata_mongodb:27017/'
    
    try:
        client = MongoClient(host_docker, serverSelectionTimeoutMS=5000)
        client.server_info() 
        db = client['previa_proyecto']
        coleccion = db['fosfato_diamonico'] # Colección específica para fosfato
        
        registros_exitosos = 0
        
        for dato in datos_finales:
            try:
                # 1. Limpieza de Precio (formato IndexMundi CLP)
                v_limpio = dato['valor_sucio'].replace('.', '').replace(',', '.')
                dato['Precio'] = float(v_limpio)
                
                # 2. Limpieza de Variación (Manejo de signos negativos y guiones)
                v_var = dato['variacion_sucia'].replace('%', '').replace(',', '.').strip()
                
                if v_var == "-" or not v_var or v_var == "0.00":
                    dato['Variacion decimal'] = 0.0
                else:
                    # El float() de Python conserva el signo "-" si existe
                    dato['Variacion decimal'] = float(v_var)
                
                # 3. Eliminar auxiliares
                del dato['valor_sucio']
                del dato['variacion_sucia']
                
                # 4. Fecha de captura al FINAL (extrema derecha)
                dato["Fecha de captura"] = time.strftime("%Y-%m-%d %H:%M:%S")
                
                coleccion.insert_one(dato)
                registros_exitosos += 1
            except:
                continue
        
        print("\n" + "="*30)
        print(f"CARGA FINALIZADA: {NOMBRE_ETIQUETA}")
        print(f"Registros guardados: {registros_exitosos}")
        print("="*30)
        
        client.close()

    except Exception as e:
        print(f"❌ ERROR DE CONEXIÓN A MONGO: {e}")
else:
    print("⚠️ No se encontraron registros para cargar.")
```
